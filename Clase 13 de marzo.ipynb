## Grafico de dispersion de puntos

## Representaci√≥n area debajo de curvas

```python
fig,ax=plt.subplots()
ax.fill_between([10,20,30,40],[20,50,60,5])
plt.show()
```

## Diagrama de barras

```python
fig,ax=plt.subplots()
ax.bar([1,2,3,4],[10,20,30,5])
plt.show()
```

## Grafico de contorno

```python
import numpy as np

fig,ax=plt.subplots()
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z=np.sqrt(x**2 + y**2)
ax.contourf(x,y,z)
plt.show()
```

## Grafico completo

```python
#Grafico mas completo

fig,ax=plt.subplots(figsize=(15,10))
dias=["L","Ma","Mi","J","V","S","D"]
temperaturas={"Medellin":[14,15,13,20,15,16,2],"Bogota":[5,6,7,10,15,6,9]}
ax.plot(dias,temperaturas["Medellin"],color="tab:purple",marker="*",linestyle="dashed",label="Medellin")
ax.plot(dias,temperaturas["Bogota"],color="tab:green",marker="o",linestyle="dotted",label="Bogota")
ax.set_title("Evolucion de la temperatura en Colombia",fontdict={"fontsize":14,"fontweight":"bold","color":"tab:red"})
ax.set_xlabel("Dias",fontdict={"fontsize":14,"fontweight":"bold","color":"tab:red"})
ax.set_ylabel("Temperaturas",fontdict={"fontsize":14,"fontweight":"bold","color":"tab:red"})
ax.grid()
ax.legend()
plt.show()
```

# Ciencias de datos

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("ggplot")
datos=pd.read_excel("winequality.xlsx")
datos
```

```python
datos.iloc[10]
```

```python
datos.fillna(0)
```

```python
rango=datos.loc[0:20]
rango
```

```python
rango.plot(x="residual sugar",y="quality",kind="bar",figsize=(15,10))
```

```python
rango1=datos.loc[datos["quality"]==8] #probar con 10 y 9
rango1
```

```python
rango1.plot(x="alcohol",y="citric acid") #Desorden
```

```python
Orden=rango1.sort_values(by=["alcohol"]) #Los ordena de forma descendente
Orden.plot(x="alcohol",y="citric acid",kind="bar",figsize=(10,10),color="black")
plt.xticks(rotation=45,horizontalalignment="center")
plt.title("Alcohol vs acido citrico")
plt.xlabel("Alcohol")
plt.ylabel("Acido citrico")
plt.show()
```

```python
Orden.plot(x="alcohol",y=["citric acid","residual sugar"],kind="bar",color=["grey","black"],figsize=(15,15))
plt.show()
```

```python
import seaborn as sns #ibreria para visualizar datos en python de forma mas profesional 

datos_cortados=datos.loc[0:200]
graf=datos_cortados.columns.to_list() #Me organiza los datos dentro de una lista pero por columna
fig,ax=plt.subplots(3,4,figsize=(15,15))
for valor,titulo in enumerate(graf):
    i=valor//4
    j=valor%4
    
    sns.histplot(x=titulo,kde=True,data=datos_cortados,ax=ax[i][j]) #kde=metodo para visualizar la distribucion
    
    ax[i][j].set_title(titulo+ "Descripcion")
    ax[i][j].set_ylabel(titulo)
plt.show()
```

```python
lista=["fixed acidity","volatile acidity","chlorides","density"]
fig,ax=plt.subplots(1,4,figsize=(15,15))
for valor,titulo in enumerate(lista):
    sns.boxplot(y=titulo,ax=ax[valor],data=datos)
plt.show()
```

```python
sns.scatterplot(x="free sulfur dioxide",y="total sulfur dioxide",data=datos)
```

```python
correlation=datos.corr(method="pearson")
correlation.head()
```

```python
fig,ax=plt.subplots(figsize=(12,12))
sns.heatmap(vmin=-1,annot=True,data=correlation,cmap="coolwarm") #annot=Metodo para visualiar los datos
plt.show()
```
